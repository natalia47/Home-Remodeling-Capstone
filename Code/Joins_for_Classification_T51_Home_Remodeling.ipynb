#Import Building Inventory Data

```python
#import building inventory dataset
bi = pd.read_csv('https://data.boston.gov/dataset/f2a82340-1f9c-49fb-a60b-b5655c9a7931/resource/391a32e6-d4bb-48d3-a990-cb35a5768a40/download/building_inventory_021020.csv', dtype={'census_tract_number':str})
bi
```

```python
#building inventory column info
bi.info(verbose=True, show_counts=True)
```

```python
#drop na in pid column for building inventory file, also drop any strings that contain alpha characters
bi=bi.dropna(subset=["pid_long"])
bi=bi[~bi["pid_long"].str.contains('[A-Za-z]',na=False)]
# set type to integer to drop float (e.g. 1000.0 becomes 1000)
bi["pid_long"]=bi["pid_long"].astype(int,errors="ignore")
#set type to string to do match
bi["pid_long"]=bi["pid_long"].astype(str)
```

```python
#rename building inventory columns with prefix bi_ to distinguish when joining
buildinginventorycolumns=bi.columns
buildinginventorycolumns=["bi_"+x for x in buildinginventorycolumns]
bi.columns=buildinginventorycolumns
```

```python
#first few rows of updated building inventory dataset with prefixes
bi.head()
```

```python
#new bi column info
bi.info(verbose=True, show_counts=True)
```
